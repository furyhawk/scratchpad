```python
from urllib.request import urlopen
import json
import pandas as pd
```

```python

uri = """https://api.bar-rts.com/replays?page=1&limit=24&hasBots=false&endedNormally=true&players=%5BTZN%5DClyret"""
```

```python
data = json.loads(urlopen(uri).read().decode("utf-8"))
```

```python
with open("data.json", "w") as outfile:
    json.dump(data['data'],outfile,indent=4, sort_keys=True)
```

```python
df = pd.json_normalize(data["data"]) #put json inside a dataframe, ["AllyTeams", "Players"]
```

```python
df_AllyTeams = df['AllyTeams'].explode().apply(pd.Series)
```

```python
df_Players = df_AllyTeams['Players'].explode().apply(pd.Series)
```

```python
# Get the winning team and count the number of wins
win_count = 0
matches = {}
for match in data['data']:
    for team in match['AllyTeams']:
        if team['winningTeam']:
            for player in team['Players']:
                if player['name'] == "[TZN]Clyret":
                    matches = {**matches, **{match['id']:match}}
```
