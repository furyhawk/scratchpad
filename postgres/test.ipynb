```python
async def get_session():
    async with database_session.get_async_session() as session:
        result = await session.execute(select(Humidity).order_by(Humidity.id))
        humiditys = result.scalars().all()
        for humidity in humiditys:
            yield humidity
```

```python
async def get_humidity(session):
    async with session() as session:
        result = await session.scalars(select(Humidity).order_by(Humidity.id))
        humidity = result.first()
        return humidity
```
